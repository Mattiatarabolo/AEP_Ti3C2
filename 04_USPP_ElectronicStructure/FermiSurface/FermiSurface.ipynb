```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline, BSpline
%matplotlib widget
```

```python
with open("bz_line.dat","r") as f:
    lines = f.readlines()
    xbz = []
    ybz = []
    for line in lines:
        splitline = line.split()
        xbz.append(float(splitline[0]))
        ybz.append(float(splitline[1]))

with open("fermi_line.dat","r") as f:
    lines = f.readlines()
    xfs = []
    yfs = []
    xfstmp = []
    yfstmp = []
    i = 0
    while i < len(lines):
        if lines[i]=="\n":
            xfs.append(np.array(xfstmp))
            yfs.append(np.array(yfstmp))
            xfstmp = []
            yfstmp = []
            i += 2
            continue
        splitline = lines[i].split()
        xfstmp.append(float(splitline[0]))
        yfstmp.append(float(splitline[1]))
        i += 1
```

```python
plt.figure(figsize=[10,10])
plt.plot(xbz,ybz,'r')
i = 0
for x in xfs:
    plt.plot(x, yfs[i], 'k')
    i += 1
#plt.ylabel("Energy [eV]")
plt.xlim((-0.7, 0.7))
plt.ylim((-0.7,0.7))
plt.axis('off')
#plt.ylim((min(energies[0])*102/100,4.8748))
#plt.ylim((-15,4.8748))
#plt.xticks(ticks = [0.0000, 0.5730, 0.9038, 1.5654], labels= ["$\Gamma$", "$M$", "$K$", "$\Gamma$"])
#plt.plot([10,10],[min(energies),max(energies)],'r--')
plt.savefig('FermiSurface.pdf')
```
